```python
from datetime import datetime

now = datetime.now()

#To Turn that object into a readable string
# %Y = Year, %m = Month, %d = Day, %H = Hour (24h), %M = Minute, %S = Second
timestamp_string = now.strftime("%Y-%m-%d_%H-%M-%S")

# To  Get specific parts if you don't want a full string
current_year = now.year
current_hour = now.hour
day_name = now.strftime("%A") # Returns "Monday", "Tuesday", etc.
```

```python
print(f"1. Object: {now}")
print(f"2. String: {timestamp_string}")
print(f"3. Specifics: Year is {current_year}, today is {day_name}")
```

```python
from datetime import datetime, timedelta

deadline = datetime(2026, 12, 31, 23, 59)
now = datetime.now()


time_remaining = deadline - now

print(f"Days left in the year: {time_remaining.days}")
print(f"Total seconds remaining: {time_remaining.total_seconds():,.0f}")

future_date = now + timedelta(days=45)
print(f"45 days from now is: {future_date.strftime('%A, %B %d')}")
```

```python
import pandas as pd
import numpy as np

# Build a DataFrame

data = {
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Chair"],
    "Price": [1200, 25, np.nan, 300, np.nan],
    "Stock": [15, 100, 45, np.nan, 20]
}

df = pd.DataFrame(data)

# INSPECT: See what the messy data looks like
print("--- ORIGINAL DATA ---")
print(df)

# 3. CLEAN: The Trapdoor (.dropna)
# This deletes any row that has even ONE missing value.
df_dropped = df.dropna()

# 4. CLEAN: The Repair (.fillna)
# Instead of deleting, we fill missing Prices with 0 and Stock with 10.
df_filled = df.copy()
df_filled["Price"] = df_filled["Price"].fillna(0)
df_filled["Stock"] = df_filled["Stock"].fillna(10)
```

```python
print("\n--- AFTER DROPNA (Rows with NaN deleted) ---")
print(df_dropped)

print("\n--- AFTER FILLNA (NaNs replaced with values) ---")
print(df_filled)
```

```python
import pandas as pd
import numpy as np

data = {
    'Time_Step': [1, 2, 3, 4, 5],
    'Temperature': [22.1, np.nan, 22.5, np.nan, 23.0],
    'Status': ['Active', np.nan, 'Active', 'Active', np.nan]
}
df = pd.DataFrame(data)

# Fillna: Replace NaNs with a specific value (like "Unknown" or 0)
df['Status'] = df['Status'].fillna("Maintenance")

# Interpolate: Smart guessing! It draws a line between 22.1 and 22.5 to find 22.3
df['Temperature'] = df['Temperature'].interpolate()

# Forward Fill (ffill): Good for categorical data where the state stays the same
# df['Column'].ffill() carries the last known value forward.

print(df)
```

```python
import pandas as pd

# Messy date strings
orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'date_str': ['2026-03-01', '05/03/2026', 'March 10, 2026']
})

orders['clean_date'] = pd.to_datetime(orders['date_str'], dayfirst=True, errors='coerce')

orders['month'] = orders['clean_date'].dt.month_name()
orders['is_weekend'] = orders['clean_date'].dt.weekday >= 5 # 5=Sat, 6=Sun

print(orders[['order_id', 'clean_date', 'month', 'is_weekend']])
```
