```python
class Robot:
    def __init__(self, name):
        self.name = name      
        self.__battery = 100
        
    def speak(self):
        return f"Hello, I am {self.name}."

r1 = Robot("R2-D2")
r2 = Robot("C-3PO")

print(r1.speak())
print(r2.speak())
```

```python
class SafeRobot(Robot):   
    def get_battery(self):
        return f"{self.name}'s battery is at {self._Robot__battery}%"
    
    def charge_battery(self, amount):
        if amount > 0 and amount <= 100:
            self._Robot__battery = amount
            return f"Charging... Battery now at {self._Robot__battery}%"
        else:
            return "Error: Invalid charge amount!"

r3 = SafeRobot("Wall-E")
print(r3.charge_battery(999))
print(r3.charge_battery(85))
```

```python
class CleaningRobot(Robot):
    def speak(self):
        return f"*Vacuum noises* I am {self.name}, the cleaner."

class SecurityRobot(Robot):
    def speak(self):
        return f"HALT! I am {self.name}. You are trespassing!"

robot_army = [
    CleaningRobot("Roomba"),
    SecurityRobot("Terminator")
]

for bot in robot_army:
    print(bot.speak())
```

```python
class BankAccount:
    def __init__(self, owner, starting_balance):
        self.owner = owner
        self.__balance = starting_balance 
        
    def get_balance(self):
        return self.__balance
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}. New balance: ${self.__balance}"
        return "Error: Cannot deposit negative money!"

    def withdraw(self, amount):
        if amount > self.__balance:
            return "Error: Insufficient funds!"
        else:
            self.__balance -= amount
            return f"Withdrew ${amount}. Remaining balance: ${self.__balance}"

alice_account = BankAccount("Alice", 500)

alice_account.__balance = 1000000 

print(alice_account.withdraw(600))
print(alice_account.deposit(200))
print(alice_account.withdraw(600))
```
