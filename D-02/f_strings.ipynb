# Day 02 - F-Strings

This prefix signals to the Python interpreter that the string is a Formatted String Literal, allowing it to evaluate expressions and inject data directly into the text at runtime.

```python
# We are assigning string data to variables in the Jupyter Kernel's memory.
user_name = "sri sai"
role = "Programmer"

# We can also assign integers (numbers).
experience_years = 2
```

```python
# Gluing text together using the + operator. 
# Notice how we have to manually add spaces (" ") and convert the number to a string using str()
old_way = "User " + user_name + " is a " + role + " with " + str(experience_years) + " years of experience."

print(old_way)
```

```python
# The 'f' activates the string. The '{}' act as portals pulling directly from memory.
# Notice how clean this is to read. No + signs, no manual spacing, no str() conversions!
modern_way = f"User {user_name} is a {role} with {experience_years} years of experience."

print(modern_way)
```

```python
# Because the {} are evaluation portals, you can do math or run functions directly inside them!
print(f"In 5 years, {user_name} will have {experience_years + 5} years of experience.")
print(f"His name in uppercase is: {user_name.upper()}")
```

```python
print(f"Isabel is {29/7:.5f} dog years old.")
# .nf here nmeans number of float values u need in the output
```

```python
print(f"""
    Most countries use the metric system for recipe measurement, 
    but American bakers use a different system. For example, they use 
    fluid ounces to measure liquids instead of milliliters (ml).
    
    So you need to convert recipe units to your local measuring system!
    
    For example, 8 fluid ounces of milk is {8 * 29.5735} ml.
    And 100ml of water is {100 / 29.5735} fluid ounces.
""")
```

```python
# Modify the code to print your age
print(f"I am {23} years old.")

# Fix this code
print(f"There are {365/7:.0f} weeks in a year")

# Complete the code
print(f"The area of a square with side 5 cm is {5*5} cm squared.")

# Modify the code to display one decimal place
print(f"The house was a good size: 1200 square feet, or {1200 * 0.092903:.1f} meters squared!")
```
