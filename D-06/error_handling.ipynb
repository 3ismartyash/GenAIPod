# Day - 06 Error handling using try catch

```python
print("1. Starting calculation...")

try:
    math_result = 10 / 0
    print("2. Math successful!") 
except ZeroDivisionError:
    print("2. Oops! You tried to divide by zero. Defaulting result to 0.")
    math_result = 0

print(f"3. Calculation finished! Result is {math_result}")
```

```python
customer = {"Name": "Sai", "Age": 25}

try:
    status = customer["Status"] 
except KeyError as e:
    print(f"ERROR: Missing data field: {e}")
    print("Applying default 'Standard' status instead.")
    status = "Standard"

print(f"Customer Status: {status}")
```

```python
print("--- OPENING SECURE DATABASE ---")

try:
    print("Reading data...")
    result = 10 / 0 
    print("Data read successfully.")
except Exception as e:
    print(f"CRITICAL FAILURE: {e}")
finally:
    print("--- CLOSING SECURE DATABASE ---")
```

```python
while True:
    try:
        age = int(input("Please enter your age: "))
        if age < 0:
            print("Age cannot be negative!")
            continue
        break
    except ValueError:
        print("Invalid input! Please enter a whole number (e.g., 25).")

print(f"Age recorded: {age}")
```

```python
try:
    val = int(input("Enter a number to square: "))
except ValueError:
    print("That wasn't a number!")
else:
    print(f"The result is {val**2}")
finally:
    print("Session complete.")
```

```python
raw_data = ["10", "20", "corrupt_data", "30", "unknown", "40"]
cleaned_data = []

print("Starting data migration...")

for item in raw_data:
    try:
        number = int(item)
        cleaned_data.append(number)
    except ValueError:
        print(f"Skipping invalid entry: '{item}'")
        continue 

print(f"Migration complete. Valid numbers found: {cleaned_data}")
```

```python
def get_user_profile(user_id, database):
    """
    Safely fetches a user. If anything goes wrong, 
    returns a 'Guest' profile instead of crashing.
    """
    try:
        user = database[user_id]
        return user
    except KeyError:
        print(f"Warning: User ID {user_id} not found.")
    except Exception as e:
        print(f"Unexpected System Error: {e}")

    return {"name": "Guest", "permissions": "Read-Only"}


mock_db = {101: {"name": "sai"}, 102: {"name": "yash"}}


print(get_user_profile(101, mock_db))

print(get_user_profile(999, mock_db))
```
