# Day 03 - lists & dictionaries

```python
# Notice the syntax: A pair of [] on the outside, containing multiple {} inside.
# Every dictionary is one "Row" of an employee database.

employee_rows = [
    {"name": "Sarah", "role": "Engineer", "salary": 90000},
    {"name": "David", "role": "Designer", "salary": 75000},
    {"name": "Maya",  "role": "Manager",  "salary": 105000}
]
```

```python
# We want to print Maya's role. Maya is the 3rd item in the list (Index 2).
# We grab index [2], and then ask for the ["role"] key.

maya_role = employee_rows[2]["role"]

print(f"Maya's role is: {maya_role}")

# We can easily loop through every row and print a dynamic f-string for each!
for row in employee_rows:
    print(f"{row['name']} works as a {row['role']}.")
```

```python
# Notice the syntax: A pair of {} on the outside. 
# The Keys are column headers. The Values are [] containing the whole column's data.

employee_columns = {
    "name": ["Sarah", "David", "Maya"],
    "role": ["Engineer", "Designer", "Manager"],
    "salary": [90000, 75000, 105000]
}
```

```python
# We want to know the total payroll (sum of all salaries).
# Because the data is column-oriented, this is trivially easy. 
# We just grab the entire "salary" list and use Python's built-in sum() function.

all_salaries = employee_columns["salary"]
total_payroll = sum(all_salaries)

print(f"The total payroll for the company is ${total_payroll:,}.")
```

```python
employees = [
    {"name": "Sarah", "role": "Engineer", "salary": 90000},
    {"name": "David", "role": "Designer", "salary": 75000},
    {"name": "Maya",  "role": "Manager",  "salary": 105000}
]
```

```python
# Goal: Find everyone who makes more than $80,000 and store them in a new list.

high_earners = [] 

for person in employees: 
    if person["salary"] > 80000:
        high_earners.append(person["name"])

print(f"High Earners: {high_earners}")
```
